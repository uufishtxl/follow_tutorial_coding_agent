# `yield`

`yield` 是 Python 中的一个关键字，用于生成器（generator）函数。可以把它看作是 `return` 的一种变体，但它不会终止函数的执行，而是返回一个值并暂停函数的执行。

## 为什么用 `yield`：节省内存

如果要处理100万条数据，用 `return` 会把所有数据都加载到内存中；而用 `yield` 只会加载一条数据。



```python
def count_up_to_n(n):
    i = 1
    while i < n:
        yield i
        i += 1


for num in count_up_to_n(3):
    print(num)
```
